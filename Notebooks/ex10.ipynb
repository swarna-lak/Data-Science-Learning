```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:/Users/ashak/OneDrive/Desktop/swarna/Superstore.csv",encoding="latin1")
print(df.head())
```

```python
sales_cat=df.groupby("Category")["Sales"].sum()
sales_cat.plot(kind="bar")
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.savefig("sales_by_category.png")
plt.show()
```

```python
profit_cat=df.groupby("Category")["Profit"].sum()
profit_cat.plot(kind="pie",autopct="%1.1f%%")
plt.title("Profit Share by Category")
plt.ylabel("")
plt.savefig("profit_pie.png")
plt.show()
```

```python
plt.hist(df["Sales"],bins=20)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.savefig("histogram.png")
plt.show()
```

```python
plt.scatter(df["Sales"],df["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.savefig("scatter.png")
plt.show()
```

```python
sns.boxplot(x="Region",y="Profit",data=df)
plt.title("Profit Distribution by region")
plt.savefig("boxpllot.png")
plt.show()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
monthly_sales=df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly_sales.plot(kind="line")
plt.title("Monthly sales Trend")
plt.ylabel("Sales")
plt.savefig("linechart.png")
plt.show()
```

```python
corr=df[["Sales","Profit","Discount","Quantity"]].corr()
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.savefig("heatmap.png")
plt.show()
```

```python
sub_sales=df.groupby("Sub-Category")["Sales"].sum()
sub_sales.sort_values().plot(kind="bar")
plt.title("Sales by Sub-Category")
plt.xlabel("Sales")
plt.savefig("horizontal_bar.png")
plt.show()
```
