```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales=[120,135,140,145,150,155,160,165,170,175,180,185,190,195,200,205,210,215,220,225]
df=pd.DataFrame({"Sales":sales})
print(df.head())
```

```python
original_mean=np.mean(sales)
print("Original Mean Sales:",round(original_mean,2))
```

```python
bootstrap_means=[]
n_iterations=1000
for i in range(n_iterations):
    sample=np.random.choice(sales,size=len(sales),replace=True)
    bootstrap_means.append(np.mean(sample))
```

```python
lower=np.percentile(bootstrap_means,2.3)
upper=np.percentile(bootstrap_means,97.5)
print("95% Confidence Interval")
print("Lower Bound:",round(lower,2))
print("Upper Bound:",round(upper,2))
```

```python
plt.hist(bootstrap_means,bins=30)
plt.axvline(lower,linestyle='-',label='Lower CI')
plt.axvline(upper,linestyle='-',label='Upper CI')
plt.axvline(original_mean,linestyle='-',label='Mean')
plt.title("Bootstap Distribution of Mean Sales")
plt.xlabel("Mean Sales")
plt.ylabel("Frequency")
plt.legend()
plt.savefig("bootstrap_confidence_interval.png")
plt.show()
```

```python
summary=pd.DataFrame({"Metric":["Mean Sales","95% CI Lower","95% CI Upper"],"Vaule":[round(original_mean,2),round(lower,2),round(upper,2)]})
print(summary)
```
