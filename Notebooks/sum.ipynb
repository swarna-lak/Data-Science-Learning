```python
import pandas as pd

# Load dataset
sales_data = pd.read_csv("C:\Users\ashak\OneDrive\Desktop\swarna\summary.csv")

print("Sales Dataset")
print(sales_data)

# Business Questions

summary = {

    "Total Sales": sales_data["Sales"].sum(),

    "Total Profit": sales_data["Profit"].sum(),

    "Average Sales": sales_data["Sales"].mean(),

    "Average Profit": sales_data["Profit"].mean(),

    "Maximum Sale": sales_data["Sales"].max(),

    "Minimum Sale": sales_data["Sales"].min(),

    "Total Quantity Sold": sales_data["Quantity"].sum(),

    "Top Product by Sales":
        sales_data.groupby("Product")["Sales"].sum().idxmax(),

    "Top Region by Sales":
        sales_data.groupby("Region")["Sales"].sum().idxmax(),

    "Total Electronics Sales":
        sales_data[sales_data["Category"]=="Electronics"]["Sales"].sum(),

    "Total Furniture Sales":
        sales_data[sales_data["Category"]=="Furniture"]["Sales"].sum(),

    "Highest Profit":
        sales_data["Profit"].max(),

    "Lowest Profit":
        sales_data["Profit"].min(),

    "Number of Orders":
        sales_data["OrderID"].count(),

    "Average Quantity":
        sales_data["Quantity"].mean()
}

# Convert summary to DataFrame
summary_df = pd.DataFrame(
    summary.items(),
    columns=["Question", "Answer"]
)

print("\nBusiness Summary")
print(summary_df)

# Export summary
summary_df.to_csv(
    "C:/Users/ashak/OneDrive/Desktop/swarna/business_summary.csv",
    index=False
)

print("\nSummary exported successfully")
```
