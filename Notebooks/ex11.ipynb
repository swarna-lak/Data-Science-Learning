```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales=[120,135,140,145,150,155,160,165,170,175]
df=pd.DataFrame({"Sales":sales})
print(df)
```

```python
mean_before=df["Sales"].mean()
median_before=df["Sales"].median()
print("Mean:",mean_before)
print("Median:",median_before)
```

```python
plt.boxplot(df["Sales"])
plt.title("SalesData Without Outliers")
plt.savefig("without_Outliers,png")
plt.show()
```

```python
sales_outlier=sales+[1000,1200]
df_outlier=pd.DataFrame({"Sales":sales_outlier})
print(df_outlier)
```

```python
mean_after=df_outlier["Sales"].mean()
median_after=df_outlier["Sales"].median()
print("Mean After Outlier:",mean_after)
print("Median After Outliers:",median_after)
```

```python
plt.boxplot(df_outlier["Sales"])
plt.title("Sales Data With Outliers")
plt.savefig("with_outliers.png")
plt.show()
```

```python
comparison=pd.DataFrame({"statistic":["Mean","Median"],"Without Outliers":[mean_before,median_before],"With Outliers":[mean_after,median_after]})
print(comparison)
```
