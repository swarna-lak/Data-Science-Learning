```python
import pandas as pd
import numpy as np
```

```python
data={
    "customerID":[101,102,103,104,105],
    "Age":[22,45,36,28,52],
    "Gender":["Male","Female","Female","Male","Female"],
    "Income":[35000,72000,55000,42000,85000],
    "Balance":[15000,35000,22000,12000,40000],
    "JoinDate":["2021-01-10",
                "2020-05-15",
                "2022-02-18",
                "2021-11-05",
                "2019-08-20"
                ],
    "Feedback":["Good service",
                "Very bad support",
                "Excellent experience",
                "Average service",
                "Poor response"
                ],
    "churn":[0,1,0,0,1]
}
df=pd.DataFrame(data)
print("Original dataset\n")
print(df)
df["SavingRatio"]=df["Balance"]/df["Income"]
df=pd.get_dummies(df,columns=["Gender"],drop_first=True)
df["JoinDate"]=pd.to_datetime(df["JoinDate"])
df["JoinYear"]=df["JoinDate"].dt.year
df["JoinMonth"]=df["JoinDate"].dt.month
df["CustomerAgeDays"]=( pd.Timestamp.today()-df["JoinDate"]).dt.days
df["Feedback_length"]=df["Feedback"].apply(len)
df["Word_count"]=df["Feedback"].apply(lambda x: len(x.split()))
df["Income_level"]=pd.cut(
    df["Income"],
    bins=[0,40000,70000,100000],
    labels=["Low","Medium","High"])
df["Income_Balance_Product"]=(df["Income"]*df["Balance"])
print("\nLeakage aware feature selection")
print("Target variable:churn")
print("CustomerID removed before model traininf because it is only identifier.")
df_model=df.drop(columns=["customerID"])
print("\nFinal feature set\n")
print(df_model)
df_model.to_csv("feature_engineered_dataset.csv",index=False)
print("\nFeature engineeres dataset saved successfully")
```
