```python
import numpy as np
scores=[]
score=float(input("Enter score:"))
while score!=-1:
    if score>=0 and score<=100:
        scores.append(score)
    else:
        print("Enter score between 0 to 100")
    score=float(input("Enter score:"))
scores=np.array(scores)
    
```

```python
print("\n summary")
print ("Total Scores:",len(scores),"\nhigh score:",np.max(scores),"\nlow score:",np.min(scores),"\naverage score",np.mean(scores),"\nmedian score:",np.median(scores))
```
