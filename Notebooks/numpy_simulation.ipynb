```python
import numpy as np
rolls=np.random.randint(1,25,10000)
expected_vaule=np.mean(rolls)
variance=np.var(rolls)
print("Expected vaule is",expected_vaule)
print("variance:",variance,"\nprobability of each outcome:")
for i in range(1,25):
    probability=np.sum(rolls==i)/10000
    print(i,":",probability)
    
```
