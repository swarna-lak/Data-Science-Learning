```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind
```

```python
np.random.seed(42)
#average spending under old campaing
group_A=np.random.normal(500,50,100)
#average spending under new campaing
group_B=np.random.normal(530,50,100)
df=pd.DataFrame({"Old campaing":group_A,
                 "New_campaing":group_B})
df.head()
```

```python
print("Old campaign Mean:",round(group_A.mean(),2))
print("New Campaign Mean:",round(group_B.mean(),2))
print("H0:Mean spending is equal in both groups")
print("H1:Mean spending differs between groups")
```

```python
t_stat,p_vaule=ttest_ind(group_A,group_B)
print("T-statistic=",round(t_stat,3))
print("P-vaule=",round(p_vaule,5))
```

```python
alpha =0.05
if p_vaule<alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
    
```

```python
if p_vaule<alpha:
    print("The new campaign produced a statistically significant change in spending.")
else:
    print("No statistically significant difference was detected.")
```

```python
import matplotlib.pyplot as plt
plt.boxplot([group_A,group_B],labels=["Old Campaign","New Campaign"])
plt.ylabel("Customer Spending")
plt.title("Comparision of Customer Spending")
plt.show()
```
