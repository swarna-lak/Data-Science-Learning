```python
import pandas as pd
from surprise import Dataset
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# -------------------------------
# Load MovieLens Dataset
# -------------------------------
data = Dataset.load_builtin("ml-100k", prompt=False)

ratings = pd.DataFrame(data.raw_ratings,
                       columns=["userId","movieId","rating","timestamp"])

print("Ratings Dataset")
print(ratings.head())

# -------------------------------
# Popularity Baseline
# -------------------------------
popular_movies = ratings.groupby("movieId")["rating"].agg(["mean","count"])

popular_movies = popular_movies.sort_values(
    by=["mean","count"],
    ascending=False
)

print("\nTop 10 Popular Movies")
print(popular_movies.head(10))

# -------------------------------
# Example Movie Metadata
# -------------------------------
movies = pd.DataFrame({
    "movieId":[1,2,3,4,5],
    "title":[
        "Toy Story",
        "GoldenEye",
        "Four Rooms",
        "Get Shorty",
        "Copycat"
    ],
    "genre":[
        "Animation Children Comedy",
        "Action Adventure Thriller",
        "Comedy",
        "Comedy Crime",
        "Crime Drama Thriller"
    ]
})

# -------------------------------
# TF-IDF
# -------------------------------
vectorizer = TfidfVectorizer(stop_words="english")

tfidf_matrix = vectorizer.fit_transform(movies["genre"])

# -------------------------------
# Cosine Similarity
# -------------------------------
similarity = cosine_similarity(tfidf_matrix)

similarity_df = pd.DataFrame(
    similarity,
    index=movies["title"],
    columns=movies["title"]
)

print("\nMovie Similarity Matrix")
print(similarity_df)

# -------------------------------
# Recommendation Function
# -------------------------------
def recommend(movie_name):

    scores = similarity_df[movie_name]

    scores = scores.sort_values(ascending=False)

    recommendations = scores.index[1:4]

    print("\nRecommended Movies for", movie_name)

    for movie in recommendations:
        print(movie)

# Example
recommend("Toy Story")

# -------------------------------
# Ranking Movies
# -------------------------------
print("\nMovie Ranking Based on Similarity")

ranking = similarity_df.sum(axis=1)

ranking = ranking.sort_values(ascending=False)

print(ranking)

# -------------------------------
# Conclusion
# -------------------------------
print("\nConcepts Covered")
print("✔ Popularity Baseline")
print("✔ Collaborative Filtering Intuition")
print("✔ TF-IDF")
print("✔ Content-Based Similarity")
print("✔ Cosine Similarity")
print("✔ Ranking Recommendations")
```
