```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/ashak/OneDrive/Desktop/swarna/Superstore.csv",encoding="latin1")
```

```python
sales_region=df.groupby("Region")["Sales"].sum()
print(sales_region)
```

```python
sales_region.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
profit_category=df.groupby("Category")["Profit"].sum()
profit_category.plot(kind="bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```
