```python
total_customers=1000
churned_customers=180
monthly_contract=400
monthly_contract_churn=120
```

```python
p_churn=churned_customers/total_customers
print("Probability of churn:",p_churn)
```

```python
p_monthly=monthly_contract/total_customers
print("Probability of Monthly Contract:",p_monthly)
```

```python
p_churn_given_monthly=monthly_contract_churn/monthly_contract
print("P(Churn| Monthly Contract):",p_churn_given_monthly)
```

```python
print("Overall churn Rate:",round(p_churn*100,2),"%")
print("Monthly Contract Churn Rate:",round(p_churn_given_monthly*100,2),"%")
```

```python
import matplotlib.pyplot as plt
categories=["Overall chuen","Monthly Contarct churn"]
values=[p_churn*100,p_churn_given_monthly*100]
plt.bar(categories,values)
plt.ylabel("Probability(%)")
plt.title("Custormer Churn Risk")
for i,v in enumerate(values):
    plt.text(i,v+1,str(round(v,1))+"%")
plt.savefig("Churn_probability.png")
plt.show()
```

```python
risk_increase=((p_churn_given_monthly-p_churn)/p_churn)*100
print(round(risk_increase,2),"%")
```
